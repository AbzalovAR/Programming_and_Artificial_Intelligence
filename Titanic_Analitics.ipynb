Есть значения 0,83 и 0,92, 0,75 - это либо ошибки где перед возрастом случайно попал нолик, либо возраст младецев, в отличие от других пассажиров на корабле записали как десятичное значение. Сомневаюсь про младецев, но попробую проверить - выведу строки с возрастом меньше 1 года:

Проверил через интернет имена - это действительно были младецы. Значит ошибки нет возраст записан верно.

Удалим строки с пустым значением Embarked:

```python
# Посмотрим уникальные значения столбца
data['Embarked'].unique()
```

```python
# Удалим пустые строки
data = data.dropna(subset=['Embarked'])
```

Теперь удалим лишние столбцы

```python
# Удалим ненужные столбцы - PassengerId тоже не несет смысла для решения нашей задачи
data.drop(['PassengerId', 'Cabin', 'Name'],axis = 1,inplace = True) # inplace = True - означает перезаписать переменную data
```

```python
data.info()
```

Переводим все строки в категории

```python
from sklearn.preprocessing import LabelEncoder

labelsToEncode=['Sex','Ticket','Embarked']
for label in labelsToEncode:
  le = LabelEncoder()
  data[label] = le.fit_transform(data[label])
```

После того как мы перевели все строчне значения в категории - мы можем проверить корреляцию:

```python
px.imshow(data.corr(),title="Корреляционная матрица")
```

Никакой явной корреляции не видно. Нас больше всего интересует столбец Survived.

```python
data.head()
```

Одним из этапов предворительной обработки данных является стандартизация значений, она используется, когда значения в большом диапазоне, для упращение тренировки модели диапазон значений уменьшается

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report,confusion_matrix
```

```python
scaler = StandardScaler()
# Нормализация числовых признаков, кроме нашего целевого признака
scaler.fit(data.drop(['Survived'],axis=1))
scaler_features = scaler.transform(data.drop(['Survived'],axis=1))
```

```python
scaler_features[0]
```

Перед тренировкой модели датасет разбивается на 2 части, на тренировочный и на тестовый датасеты. Это делается для того, что бы после окончания тренировки модели проверить ее точность на данных, которая модель не видела

```python
from sklearn.model_selection import train_test_split
# X = scaler_features
X = data.drop(['Survived'],axis=1)
Y = data['Survived']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state=4)
```

Ниже предоставленны несколько моделей и их метрики, для определения точности модели использовались следующие метрики


1.   Среднеквадратичная ошибка
2.   Среднеквадратичная ошибка в процентах

Линейная регрессия

```python
# Среднее арифметическое  квадратов разностей между предсказанными и реальными значениями:
from sklearn.metrics import mean_squared_error

# относительная ошибка, где разность между наблюдаемым и фактическим значениями делится на наблюдаемое значение и выражается в процентах:
from sklearn.metrics import mean_absolute_percentage_error

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LassoCV
```

```python
reg=LinearRegression()
reg.fit(X_train,Y_train)
```

```python
pred = reg.predict(X_test)
```

```python
X_test
```

```python
pred
```

```python
Y_test
```

```python
mean_squared_error(Y_test, pred)
```

```python
mean_absolute_percentage_error(Y_test, pred)
```

ЧТО это означает? Невероятно огромный процент не точности!

Случайный лес

```python
regr = RandomForestRegressor(max_depth=4, random_state=4)
regr.fit(X_train, Y_train)
```

```python
pred2 = regr.predict(X_test)
```

```python
mean_squared_error(Y_test, pred2)
```

```python
mean_absolute_percentage_error(Y_test, pred2)
```

```python
lassoReg = LassoCV(cv=5, random_state=0).fit(X_train, Y_train)
```

```python
pred3 = lassoReg.predict(X_test)
```

```python
mean_squared_error(Y_test, pred3)
```

```python
mean_absolute_percentage_error(Y_test, pred3)
```
